```
import scipy.integrate as integrate
import numpy as np
from scipy.integrate import quad, dblquad, simps, trapz
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

One variable integral

```
def f(x):
  return np.exp(x) +x**2
i=quad(f,1,2)
print(i)
plt.plot(x,f(x))
plt.show()
```

An integral with a parameter


```
def f(x,a,b):
  return np.exp(x)+a*x+b
a=1
b=1
I=quad(f,0,1,args=(a,b))
print(I)
plt.plot(x,f(x,1,1))
plt.show()
```

Dobule integral


```
f= lambda x,y: x**2+y
x=np.linspace(-2,2,20)
y=np.linspace(-2,2,20)
X,Y=np.meshgrid(x,y)
F=f(X,Y)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,F)

res=dblquad(f,0,1, lambda x:0 ,lambda x:1)
print(res)
```

An integral of sampels


```
def f(x):
  return x**2
x=np.array([1,3,2,5,4])
s1=np.sort(x)
res=simps(f(x),s1)
plt.plot(s1,f(x))
plt.show()
print(res)
```
