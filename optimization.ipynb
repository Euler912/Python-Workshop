```
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from scipy.optimize import shgo
from scipy import linalg
```

Finding the min of scalar function 

```
f=lambda x: x**4-3*x**2+2
x=np.linspace(-3,3,30)
plt.plot(x,f(x))
res=optimize.minimize_scalar(f)
print(res.x,f(res.x))
```

Finding the min of constrained function (bounded):

```
f=lambda x: x**4-3*x**2+2
x=np.linspace(-3,3,30)
y=f(x)
plt.plot(x,y)
res1 = optimize.fminbound(f, -3, 0)
print(res1,f(res1))
```

Finding the max of a  scalar function :


```
g=lambda x:-1* (x**4-3*x**2+2)
x=np.linspace(-3,3,30)
y=-1*g(x)
plt.plot(x,y)
res=optimize.minimize_scalar(g)
print(res.x,(-1)*g(res.x))
```

finding root a scalar function

```
def f(x): 
  return x**2+np.exp(x)-6
x=np.linspace(-3,3,50)
plt.plot(x,f(x))
root=optimize.brentq(f,0,2)
print (root,f(root))
```

finding root multi vari function without jacobian:

```
def func(x):
  # we used np array because idan recommended to use it when we use libs that relate to numpy 
  f=np.array([x[0]**2*x[1]+np.cos(x[0]), x[1]*x[0]-x[1]+1])
  return f
res=optimize.root(func,[0,0])
print(res)
#plot of the graphs: f1 and f2 
f1= lambda x,y : (x**2)*y+np.cos(x)
x=np.linspace(-3,3,30)
y=np.linspace(-3,3,30)
X,Y= np.meshgrid(x,y)
Z=f1(X,Y)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)
f2=lambda x, y: x*y-y+1
T=f2(X,Y)
ax.plot_surface(X,Y,T)
```

Finding root multi vari function with the jacobian

```
def f(x):
    f=np.array([x[0]**2*x[1]+np.cos(x[0]),  x[1]*x[0]-x[1]+1])
    df=np.array([[2*x[0]*x[1]-np.sin(x[0]),x[0]**2], [x[1],x[0]-1]])
    return f ,df
res1=optimize.root(f,[0,0],jac=True)
print(res1)
#plot of the graphs: f1 and f2 
f1= lambda x,y : (x**2)*y+np.cos(x)
x=np.linspace(-3,3,30)
y=np.linspace(-3,3,30)
X,Y= np.meshgrid(x,y)
Z=f1(X,Y)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)
f2=lambda x, y: x*y-y+1
T=f2(X,Y)
ax.plot_surface(X,Y,T)
```

min of a multi variables function

```
def g(x):
  return np.array(x[0]*x[1]+x[0]**2+np.cos(x[0]))
min_func=optimize.minimize(g,[1,1],method='SLSQP')
print(min_func.x)
#plotting the graph:
g2=lambda x,y : x*y+x**2*np.cos(x)
X,Y= np.meshgrid(x,y)
Z=g2(X,Y)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)
```

Max of a multi variables function

```
def h(x):
  return np.array(-1*(x[0]*x[1]+x[0]**2+np.cos(x[0])))
max_func=optimize.minimize(h,[1,1],method='SLSQP')
print((-1)*max_func.x)
#plotting the graph:
h2=lambda x,y :(x*y+x**2*np.cos(x))
X,Y= np.meshgrid(x,y)
Z=h2(X,Y)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)
```
